```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import draw
import random
```

```python
def circle(arr, n, r):
    arr_shape = arr.shape
    for i in range(n):
        x = random.randint(0,min(arr_shape)-r)
        y = random.randint(0, max(arr_shape)-r)
        
        row, col = draw.disk((x,y), r)
        arr[row,col] = 1
        
    return arr
```

```python
img = np.zeros((5*10**3,8*10**3))
```

```python
plt.imshow(img)
```

```python
star_img = circle(img, 3*10**4, 5)
plt.imshow(star_img)
plt.imsave('imgh.jpg', star_img)
```
