### Used to generate random dots in an image

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import draw
import random
```

```python
def circle(arr, n):
    arr_shape = arr.shape
    for i in range(n):
        x = random.randint(0,arr_shape[0]-1)
        y = random.randint(0, arr_shape[1]-1)
        intensity = random.randint(0,255)
        
#         row, col = draw.disk((x,y), r)
        arr[x,y] = intensity
        
    return arr
```

### Enter Size of image below:

```python
img = np.zeros((3000, 3000)) # Dimension of the Image
```

### Kernel may die......

```python
star_img = circle(img, 10**4)
plt.imshow(star_img, cmap="gray")
plt.imsave('imgh.jpg', star_img, cmap="gray")
```
