### In this notebook I am trying to generate fake star constillations for training the model

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def star(arr, n, r):
    for i in range(n):
        x = np.random.randint(0, min(arr.shape)-r, 1)
        y = np.random.randint(0, max(arr.shape)-r, 1)

        arr[(x,y)] = 1
        arr[(x-r,y)] = 1
        arr[(x+r,y)] = 1
        arr[(x,y-r)] = 1
        arr[(x,y+r)] = 1
        arr[(x+r,y+r)] = 1
        arr[(x-r,y+r)] = 1
        arr[(x+r,y-r)] = 1
        arr[(x-r,y-r)] = 1
    
    return arr
```

```python
def star2(arr, n, r):
    for i in range(n):
        x = np.random.randint(0, min(arr.shape)-r, 1)
        y = np.random.randint(0, max(arr.shape)-r, 1)
        
        xx, yy = np.mgrid[:r, :r]
        circle = (xx - 100) ** 2 + (yy - 100) ** 2
        donut = np.logical_and(circle < (10000), circle+1)
        
    return donut
```
