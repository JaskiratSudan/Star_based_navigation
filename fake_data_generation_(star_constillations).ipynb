### In this notebook I am trying to generate fake star constillations for training the model

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
img1 = np.zeros((28,28))
```

```python
n=6
index_one = np.random.choice(img1.size, n, replace=False)
img1.ravel()[index_one] = 1
```

```python
plt.imshow(img1, cmap='gray')
```

```python
img_arr = np.empty((0,28,28), int)
for i in range(10):
    img = np.zeros((28,28))
    n = np.random.randint(10)
    index_one = np.random.choice(img.size, n, replace=False)
    img.ravel()[index_one] = 1
    img_arr = np.append(img_arr, [img], axis=0)
```

```python
fig, ax = plt.subplots(2,5, figsize=(10,7))
ax = ax.ravel()
for i in range(10):
    ax[i].imshow(img_arr[i], cmap='gray')
    plt.tight_layout()
```
